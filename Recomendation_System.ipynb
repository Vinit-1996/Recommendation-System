## Build a recommender system by using cosine simillarties score.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine , correlation
```

```python
books=pd.read_csv("/content/book.csv",encoding='latin 1',index_col=0)
books.head()
```

```python
books.rename({"User.ID":"userid","Book.Title":'title',"Book.Rating":"rating"},axis=1,inplace=True)
books.head()
```

```python
books.shape
```

```python
books.isnull().sum()
```

```python
len(books["userid"].unique())
```

```python
len(books["title"].unique())
```

```python
books.duplicated().sum()
```

```python
books.drop_duplicates(inplace=True)
```

```python
books_df= books.pivot_table(index="userid",columns="title",values='rating').reset_index(drop=True)
```

```python
books_df
```

```python
books_df.index=books.userid.unique()
```

```python
books_df
```

```python
books_df.fillna(0,inplace=True)
```

```python
books_df.index
```

```python
books_sim= 1- pairwise_distances(books_df.values,metric='cosine')
```

```python
books_sim
```

```python
books_sim_df=pd.DataFrame(books_sim)
books_sim_df
```

```python
books_sim_df.index=books.userid.unique()
books_sim_df.columns=books.userid.unique()
```

```python
books_sim_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(books_sim,0)
books_sim_df.iloc[0:5,0:5]
```
